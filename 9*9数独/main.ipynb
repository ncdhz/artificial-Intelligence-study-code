```python
import tkinter as tk
import copy
```

```python
class Sudoku:
    def __init__(self):
        self.window = tk.Tk()
        self.window.title('9*9数独')
        self.row = 0
        self.column = 0
        # 用于给frame添加滚动条
        f = tk.Frame(self.window)
        self.canvas = tk.Canvas(f, height=500, width=1060)
        sb = tk.Scrollbar(f, orient=tk.VERTICAL, command=self.canvas.yview)
        self.canvas.config(yscrollcommand=sb.set)
        f.pack(fill=tk.BOTH, expand=True)
        sb.pack(side=tk.RIGHT, fill=tk.Y)
        self.canvas.pack(side=tk.RIGHT, fill=tk.BOTH, expand=True)

    def draw(self, datas):
        # 此 frame 中有滚动条
        frame = tk.Frame(self.window)
        self.canvas.create_window(0, 0, anchor=tk.NW, window=frame)
        # 画数独图
        for i in range(len(datas)):
            canvas = tk.Canvas(frame, bg='green', height=290, width=290)
            for j in range(10):
                if j % 3 == 0:
                    canvas.create_line(10 + j * 30, 10, 10 + j * 30, 280, width=3)
                    canvas.create_line(10, 10 + j * 30, 280, 10 + j * 30, width=3)
                else:
                    canvas.create_line(10 + j * 30, 10, 10 + j * 30, 280)
                    canvas.create_line(10, 10 + j * 30, 280, 10 + j * 30)
            
            for j in range(len(datas[i])):
                for k in range(len(datas[i][j])):
                    canvas.create_text(25 + k * 30, 25 + 30 * j, text=str(datas[i][j][k]))
            
            # 排列canvas
            canvas.grid(row = self.row, column = self.column)
            self.column = self.column + 1
            if self.column >= 4:
                self.column %= 4
                self.row += 1
        self.canvas.configure(scrollregion=(0, 0, 0, 300 * len(datas) // 4))

    def run(self):
        def test_rule(d, i, j):
            '''
            9*9数独问题中的约束
            用于检测某个节点是否符合约束
            i 行
            j 列
            '''
            for k in range(9):
                if k != j and d[i][k] == d[i][j]:
                    return False
                if k != i and d[k][j] == d[i][j]:
                    return False
            '''
            用于判断如 1 所示这样的格子中不能出现重复的数字
            |-------|-------|-------|
            |       |       |       |
            |   1   |       |       |
            |       |       |       |
            |-------|-------|-------|
            |       |       |       |
            |       |       |       |
            |       |       |       |
            |-------|-------|-------|
            |       |       |       |
            |       |       |       |
            |       |       |       |
            |-------|-------|-------|
            '''
            for x in range(3):
                for y in range(3):
                    save_data = []
                    for m in range(3):
                        for n in range(3):
                            if d[m + x * 3][n + y * 3] == 0:
                                continue
                            if d[m + x * 3][n + y * 3] in save_data:
                                return False
                            save_data.append(d[m + x * 3][n + y * 3])
            return True
        def find_zero(d):
            for i in range(len(d)):
                for j in range(len(d[i])):
                    if d[i][j] == 0:
                        return [i, j]
            return False
        data = [
            [0, 2, 0, 5, 0, 1, 0, 9, 0],
            [8, 0, 0, 2, 0, 3, 0, 0, 6],
            [0, 3, 0, 0, 6, 0, 0, 7, 0],
            [0, 0, 1, 0, 0, 0, 6, 0, 0],
            [5, 4, 0, 0, 0, 0, 0, 1, 9],
            [0, 0, 2, 0, 0, 0, 7, 0, 0],
            [0, 9, 0, 0, 3, 0, 0, 8, 0],
            [2, 0, 0, 8, 0, 4, 0, 0, 7],
            [0, 1, 0, 9, 0, 7, 0, 6, 0]
        ]

        # 用于存储数独所有解
        datas = [data]
        def sudoku(d, i, j):
            for x in range(9):
                d[i][j] = x + 1
                if test_rule(d, i, j):
                    if find_zero(d):
                        m, n = find_zero(d)
                        sudoku(d, m, n)
                    else:
                        datas.append(copy.deepcopy(d))
                d[i][j] = 0
        i, j = find_zero(data)
        sudoku(data, i, j)
        return datas
    def show(self):
        self.window.mainloop()
```

```python
s = Sudoku()
datas = s.run()
s.draw(datas)
s.show()
```
