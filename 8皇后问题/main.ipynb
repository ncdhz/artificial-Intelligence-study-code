```python
import tkinter as tk
import copy
```

```python
class Queen8:
    def __init__(self):
        self.window = tk.Tk()
        self.window.title('8皇后问题')
        self.row = 0
        self.column = 0
        # 用于给frame添加滚动条
        f = tk.Frame(self.window)
        self.canvas = tk.Canvas(f, height=500, width=1060)
        sb = tk.Scrollbar(f, orient=tk.VERTICAL, command=self.canvas.yview)
        self.canvas.config(yscrollcommand=sb.set)
        f.pack(fill=tk.BOTH, expand=True)
        sb.pack(side=tk.RIGHT, fill=tk.Y)
        self.canvas.pack(side=tk.RIGHT, fill=tk.BOTH, expand=True)
        
    def draw(self, datas):
        # 此 frame 中有滚动条
        frame = tk.Frame(self.window)
        self.canvas.create_window(0, 0, anchor=tk.NW, window=frame)
        # 画 8 皇后问题图
        for i in range(len(datas)):
            canvas = tk.Canvas(frame, bg='green', height=260, width=260)
            for j in range(9):
                canvas.create_line(10 + j * 30, 10, 10 + j * 30, 250)
                canvas.create_line(10, 10 + j * 30, 250, 10 + j * 30)
            for j in range(len(datas[i])):
                for k in range(len(datas[i][j])):
                    if datas[i][j][k] == 1:
                        canvas.create_oval(10 + k * 30, 10 + 30 * j, 10 + (k + 1) * 30, 10 + 30 * (j + 1))
            # 排列canvas
            canvas.grid(row = self.row, column = self.column)
            self.column = self.column + 1
            if self.column >= 4:
                self.column %= 4
                self.row += 1
        self.canvas.configure(scrollregion=(0, 0, 0, 270 * len(datas) // 4))
    
    def run(self):
        def test_rule(d, i, j):
            '''
            8皇后问题中的约束
            用于检测某个节点是否符合约束
            i 行
            j 列
            '''
            # 用于检测一列中是否存在元素
            for ii in range(i):
                if d[ii][j] == 1:
                    return False
            
            # 用于检测左斜线是否存在元素
            for x in range(i):
                if j - x - 1 < 0:
                    break
                if d[i - x - 1][j - x - 1] == 1:
                    return False
            # 用于测试右斜线是否存在元素
            for x in range(7 - j):
                if i - x - 1 < 0:
                    return True
                if d[i - x - 1][j + x + 1] == 1:
                    return False
            return True
        # 存放所有满足 8 皇后问题的数组
        datas = []
        # 通过回溯法求解 8 皇后问题
        def queen8(d, i, j):
            # 判断i，j位置下子是否符合8皇后问题规则
            if not test_rule(d, i, j):
                return
            if i == 7:
                d[i][j] = 1
                datas.append(copy.deepcopy(d))
                d[i][j] = 0
                return
            for jj in range(8):
                d[i][j] = 1
                queen8(d, i + 1, jj)
                d[i][j] = 0
        
        data = [[0 for x in range(8)] for y in range(8)]
        for j in range(8):
            queen8(data, 0, j)
        return datas
    def show(self):
        self.window.mainloop()
q = Queen8()
datas = q.run()
q.draw(datas)
q.show()
```
