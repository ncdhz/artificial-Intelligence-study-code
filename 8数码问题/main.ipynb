```python
import tkinter as tk
import copy
```

```python
class Draw8:
    def __init__(self):
        self.window = tk.Tk()
        self.window.title('8数码问题')
        self.row = 0
        self.column = 0
    def draw(self, data, now = False, original = False):
        canvas = tk.Canvas(self.window, bg='green', height=170, width=170)
        for i in range(4):
            canvas.create_line(10 + i * 50, 10, 10 + i * 50, 160)
            canvas.create_line(10, 10 + i * 50, 160, 10 + i * 50)
        for i in range(len(data)):
            for j in range(len(data[i])):
                if data[i][j]:
                    canvas.create_text(j * 50 + 35, 35 + i * 50, text=str(data[i][j]))   
        if now:
            canvas.create_rectangle(15 + now[1] * 50, 15 + now[0] * 50, 55  + now[1] * 50, 55 + now[0] * 50, outline='red')
        if original:
            canvas.create_rectangle(15 + original[1] * 50, 15 + original[0] * 50, 55  + original[1] * 50, 55 + original[0] * 50, outline='blue')
            canvas.create_line( 35  + now[1] * 50, 35 + now[0] * 50,35 + original[1] * 50, 35 + original[0] * 50, arrow=tk.LAST)
        # 用于调整布局
        canvas.grid(row = self.row, column = self.column)
        self.column = self.column + 1
        if self.column >= 4:
            self.column %= 4
            self.row += 1
    def show(self):
        self.window.mainloop()
```

```python
class Digital8:
    def __init__(self, data):
        self.data = data
    def isOK(self):
        list = []
        isOk = 0
        for i in range(len(self.data)):
            for j in range(len(self.data[i])):
                if self.data[i][j]:
                    for k in range(len(list)):
                        if list[k] > self.data[i][j]:
                            isOk += 1
                    list.append(self.data[i][j])
                else:
                    self.x = i
                    self.y = j
        return True if isOk % 2 == 0 else False
    def distance(self, data = False):
        if not data:
            data = self.data
        sum = 0
        for i in range(len(data)):
            for j in range(len(data[i])):
                if data[i][j]:
                    sum += abs(data[i][j] // 3 - i) + abs(data[i][j] % 3 - j)
        return sum
    def solve(self):
        def listFind(l, data):
            for i in range(len(l)):
                if data == l[i][0]:
                    return True
            return False
        close_list = []
        open_list = []
        if self.isOK():
            open_list.append([self.data, self.distance() , self.x, self.y, 0, False])
            while len(open_list):
                min = 0
                for i in range(len(open_list)):
                    if open_list[min][1] > open_list[i][1]:
                        min = i
                d = open_list.pop(min)
                close_list.append(d)
      
                dd, _, x, y, deep, _ = copy.deepcopy(d)
                
                ddd = copy.deepcopy(dd)
                
                for l in [[0, -1], [0, 1], [1, 0], [-1, 0]]:
                    if x + l[0] >= 0 and x + l[0] < 3 and y+ l[1] >= 0 and y + l[1] < 3:
                        ddd[x][y] = ddd[x + l[0]][y + l[1]]
                        ddd[x + l[0]][y + l[1]] = False
                        if not listFind(close_list, ddd):
                            if self.distance(ddd) == 0:
                                return [ddd, self.distance(ddd) + deep + 1, x + l[0], y + l[1], deep + 1, d]
                            open_list.append([ddd, self.distance(ddd) + deep + 1, x + l[0], y + l[1], deep + 1, d])
                        ddd = copy.deepcopy(dd)
        else:
            print('没有解')
        return False
startState = [[0,4,2], [1,6,5], [7,3,8]]
d8 = Digital8(startState)
solve_path = d8.solve()
d = Draw8()
list = []

if solve_path:
    while solve_path[5]:
        list.append([solve_path[0], solve_path[2], solve_path[3]])
        solve_path = solve_path[5]
    list.append([solve_path[0], solve_path[2], solve_path[3]])
original = []
while len(list):
    now = list.pop()
    if len(original):
        d.draw(now[0], [now[1], now[2]], original)
    else:
        d.draw(now[0], [now[1], now[2]])
    original = [now[1], now[2]]
d.show()
```
